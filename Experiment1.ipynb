```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

number_str = input("Enter a number to compute its factorial: ")
number = int(number_str)
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")

number_to_display_str = input("Enter a number to display its multiplication table: ")
number_to_display = int(number_to_display_str)
display_multiplication_table(number_to_display)
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

num_input_str = input("Enter a number to check if it's odd or even: ")
num_input = int(num_input_str)
print(check_odd_even(num_input))
```

```python
def is_palindrome_number(number):
    # Convert the number to a string for easy reversal
    num_str = str(number)
    # Compare the original string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome number."
    else:
        return f"{number} is not a palindrome number."

num_input_str = input("Enter a number to check if it's a palindrome: ")
num_input = int(num_input_str)
print(is_palindrome_number(num_input))
```

```python
def recommend_stream(subject):
    subject_lower = subject.lower()
    if "math" in subject_lower or "physics" in subject_lower or "chemistry" in subject_lower :
        return "Based on your interest in Math, Science, and Engineering, a Science stream (e.g., Computer Engineering) would be a great fit!"
    elif "biology" in subject_lower or "zoology" in subject_lower or "botany" in subject_lower:
        return "Given your passion for Biology, a Medical/Life Sciences stream (e.g., Medicine, Biotechnology) seems ideal."
    elif "history" in subject_lower or "geography" in subject_lower or "literature" in subject_lower or "political science" in subject_lower:
        return "With your focus on Humanities subjects, an Arts/Humanities stream (e.g., Law, Journalism, Social Sciences) would be beneficial."
    elif "business" in subject_lower or "economics" in subject_lower or "accounting" in subject_lower:
        return "Your inclination towards Business and Finance suggests a Commerce stream (e.g., Business Administration, Accounting, Finance)."
    else:
        return "I can't recommend a specific stream based on that subject. Please try entering a more common academic subject like Math, Biology, History, or Business."

# Get user input
user_subject = input("Enter a subject you are most interested in (e.g., Math, Biology, History, Business): ")

# Get and print the recommendation
recommendation = recommend_stream(user_subject)
print(recommendation)
```
